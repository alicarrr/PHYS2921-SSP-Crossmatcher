```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
from astropy.coordinates import SkyCoord, Distance
import astropy.units as un
from astropy.coordinates import Angle, Latitude, Longitude

df_123 = pd.read_csv('Hot_Jupiters_123_Samples_proper_corrected_NASA.csv')
coords_123 = SkyCoord(ra = df_123['ra'] * un.deg, dec = df_123['dec'] * un.deg, unit = 'deg')
corrected_coords_123 = SkyCoord(ra = df_123['ra_corrected'] * un.deg, dec = df_123['dec_corrected'] * un.deg, unit = 'deg')
correction_123 = coords_123.separation(corrected_coords_123)
correction_arcsec_123 = correction_123.arcsecond

correction_arcsec_123
       
```

```python
correction_new_123 = correction_arcsec_123[~np.isnan(correction_123)]
correction_new_123

#plt.hist(correction_new_123, bins = 8)
plt.hist(correction_new_123, bins = 10)
#plt.xlim(0,correction_new_123.max())
#plt.xticks(0,correction_new_123.max())
plt.title('Proper motion correction size (123 samples)')
plt.xlabel('Separation (arcsec)')
plt.show()
```

```python
#With updated bins 
correction_new_123 = correction_arcsec_123[~np.isnan(correction_123)]
correction_new_123
xbins = np.arange(0,50,5)

plt.hist(correction_new_123, bins = xbins, align = 'left')
plt.title('Proper motion correction size (123 samples)')
plt.xlabel('Separation (arcsec)')
plt.show()
```

```python
df = pd.read_csv('Hot_Jupiters_10_Samples_proper_corrected_NASA.csv')
coords = SkyCoord(ra = df['ra'] * un.deg, dec = df['dec'] * un.deg, unit = 'deg')
corrected_coords = SkyCoord(ra = df['ra_corrected'] * un.deg, dec = df['dec_corrected'] * un.deg, unit = 'deg')
correction = coords.separation(corrected_coords)
correction_arcsec = correction.arcsecond

correction_arcsec
```
